# 5-Visualize the gender, age, award category, and years of Nobel Prize winners.

```python
data['birth_date'].dtype
```

```python
data['birth_date']=pd.to_datetime(data['birth_date'])
data['birth_date']
```

```python
data['year']=data['year'].astype('float64')
data['year']
```

```python
data['birth_date'] = pd.to_datetime(data['birth_date']).dt.year
data['birth_date']
```

```python
data['age']=data['year']-data['birth_date']
```

```python
data['age']
```

```python
data
```

```python
data['category'].unique()
```

```python
for category in data['category'].unique():
    category_data = data[data['category'] == category]
    
    plt.figure(figsize=(6,4))
    plt.scatter(category_data[category_data['sex'] == 'Male']['year'], category_data[category_data['sex'] == 'Male']['age'], label='Male', color='blue')
    plt.scatter(category_data[category_data['sex'] == 'Female']['year'], category_data[category_data['sex'] == 'Female']['age'], label='Female', color='red')
    plt.title(f'Ages of Nobel Prize Winners Over Years by Sex in {category}')
    plt.xlabel('Year')
    plt.ylabel('Age')
    plt.legend()
    plt.grid(True)
    plt.show()


```

# 6-Visualize the categories and countries of Nobel Prize winners between 1938 and 1945.

```python
data_btw=data[(data['year']>=1938) & (data['year']<=1945)]
```

```python
category_btw=data_btw['category'].value_counts()
```

```python
countries_btw=data_btw['birth_country'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
category_btw.plot(kind='bar', color='skyblue')
plt.title('Nobel Prize Winners by Category (1938-1945)')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(countries_btw, labels=countries_btw.index, autopct='%1.1f%%', startangle=140)
plt.title('Nobel Prize Winners by Country (1938-1945)')
plt.axis('equal');
```

# 7-Visualize the categories and countries of Nobel Prize winners between 1947 and 1991.

```python
filtered_data = data[(data['year'] >= 1947) & (data['year'] <= 1991)]

categories = filtered_data['category'].unique()
for category in categories:
    category_data = filtered_data[filtered_data['category'] == category]
   
    country_counts = category_data['birth_country'].value_counts()
    
    plt.figure(figsize=(6, 4))
    plt.bar(country_counts.index, country_counts.values, color='purple')
    plt.title(f'Nobel Prize Winners in {category} Category (1947-1991)')
    plt.xlabel('Country')
    plt.ylabel('Number of Winners')
    plt.xticks(rotation=90)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

```

```python
    plt.figure(figsize=(10, 8))
    plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Nobel Prize Winners by Country (1947-1991)')
    plt.axis('equal')
    plt.show()
```
